```python
# Read in spss dataset and put into variable df_mipd
# df is for dataframe

df_mipd = pd.read_spss('data/MIPD in spss.sav')
```

```python
# get infor of the dataset

df_mipd.info()
```

```python
# again get the column headers

df_mipd.columns.to_list()
```

```python
# create a subset of the data using a select list of columns

sub_mipd = df_mipd[["studyid", "fw_start", "samplesize",
                    "party_id", "mip_cap1",  "mip_singer1", "mip_marpor1"]]
```

```python
# display subset 
sub_mipd
```

```python
# using group_by to group columns in order 

freq_study = sub_mipd.groupby(["studyid","party_id","mip_cap1"])
```

```python
# Looking at a set studyid from the subset 

sub_mipd[sub_mipd["studyid"] == 'ANES-1960']
```

```python
# Loop through the studyids and getting a count of responses for each political group
# limited to the first 10 studyids

for study in sub_mipd["studyid"].unique()[:10]:
    print(study)
    temp_df = sub_mipd[sub_mipd["studyid"] == study]
    print(temp_df["party_id"].value_counts())
```

```python
# Brute force way to get column counts for particular group_bys


sub_mipd.groupby(["studyid","party_id", 'mip_cap1'])['mip_cap1'].count()[:50]

# sub_mipd.groupby(["studyid","party_id", 'mip_singer1'])['mip_singer1'].count()[:60]

# sub_mipd.groupby(["studyid","party_id", 'mip_marpor1'])['mip_marpor1'].count()[:60]
```

```python
# better and less problematic method of unstacking grouped data
# and getting counts

freq_study = sub_mipd.groupby(["studyid","party_id","mip_cap1"])
mip_cap1_counts = freq_study['mip_cap1'].count().unstack()
# mip_singer1_counts = freq_study['mip_singer1'].count().unstack()
# mip_marpor1_counts = freq_study['mip_marpor1'].count().unstack()
```

```python
mip_cap1_counts
```

```python
# save cap1 counts to file 

mip_cap1_counts.to_csv('output/mip_capi1_counts.csv')
```

```python
freq_study_singer = sub_mipd.groupby(["studyid","party_id","mip_singer1"])
mip_singer1_counts = freq_study_singer['mip_singer1'].count().unstack()
mip_singer1_counts
```

```python
# save singer sounts to file

mip_singer1_counts.to_csv("output/mip_singer1_counts.csv")
```

```python
freq_study_marpor = sub_mipd.groupby(["studyid","party_id","mip_marpor1"])
mip_marpor1_counts = freq_study_marpor['mip_marpor1'].count().unstack()
mip_marpor1_counts
```

```python
# save marpor counts to file

mip_marpor1_counts.to_csv("output/mip_marpor1_counts.csv")
```
